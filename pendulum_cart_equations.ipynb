```python
# Import sympy package
import sympy as sp
from IPython.display import display, Math
```

```python
# Define the physical variables
t       = sp.Symbol('t', real=True)
mp      = sp.Symbol('m_p', positive=True)
mc      = sp.Symbol('m_c', positive=True)
g       = sp.Symbol('g', positive=True)
l       = sp.Symbol('l', positive=True)
theta   = sp.Function('theta')(t)
x       = sp.Function('x')(t)


# Cartesian coordinates of pendulum bob
x_pend = x + l * sp.sin(theta)
y_pend = - l * sp.cos(theta)

# Position vector of the pendulum bob
r = sp.Matrix([x_pend, y_pend])

# Velocity vector of the pendulum bob
v = r.diff(t)

print("\nVelocity vector:")
display(v)

# --- Kinetic energy ---
T_cart = sp.Rational(1,2) * mc * x.diff(t)**2
T_pend = sp.Rational(1,2) * mp * (v.dot(v))
T      = sp.simplify(T_cart + T_pend)

# --- Potential energy ---
V = mp * g * y_pend
V = sp.simplify(V)

# --- Lagrangian ---
L = T - V
L = sp.simplify(L)

# Display results

print("\nKinetic Energy T cart:")
display(T_cart)

print("\nKinetic Energy T pendulum:")
display(T_pend)

print("\nKinetic Energy T:")
display(T)

print("\nPotential Energy V:")
display(V)

print("\nLagrangian L = T - V:")
display(L)
```

```python
# EulerLagrange for x
EL_x = sp.simplify(sp.diff(sp.diff(L, x.diff(t)), t) - sp.diff(L, x))

display(Math(r'\text{Equation in } x:'))
print('')
display(Math(sp.latex(EL_x)))


# EulerLagrange for theta
EL_theta = sp.simplify(sp.diff(sp.diff(L, theta.diff(t)), t) - sp.diff(L, theta))

display(Math(r'\text{Equation in } \theta:'))
print('')
display(Math(sp.latex(EL_theta)))





```
